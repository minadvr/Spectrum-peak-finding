```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn.model_selection import train_test_split
import seaborn as sns
from sklearn import metrics
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
a=pd.read_fwf('data.txt')  
x_list=[]
y_list=[]

for i in range(len(a)//8):
    x=float(a['data'][3+(8*i)][11:-11])
    x_list.append(x)
    y=float(a['data'][5+(8*i)][11:-11])
    y_list.append(y)
    
    
```

```python
df = pd.DataFrame(list(zip(x_list, y_list)), 
               columns =['x', 'y'])
```

```python
df.head()
```

```python
sns.jointplot(x=df['x'], y=df['y'], data=df, kind='reg')
```

```python
plt.plot(x_list,y_list)
```

```python
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
peaks, _ = find_peaks(y_list, height=2000)
plt.plot(y_list)
plt.plot(peaks, df['y'][peaks], "x")
plt.plot(np.zeros_like(y_list), "--", color="gray")
plt.show()
```
